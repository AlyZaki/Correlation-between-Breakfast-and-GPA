```python
df = pd.read_csv("aly.csv")
```

```python
df.describe()
```

```python

sns.regplot(x='CGPA', y='breakfast days', data=df, scatter_kws={'alpha':0.7})
plt.title('Relationship between Breakfast and CGPA')
plt.xlabel('Breakfast days')
plt.ylabel('CGPA')
plt.grid(True)
plt.show()
```

```python
df['breakfast days'].corr(df['CGPA'])
```

```python

categories = ['Yes', 'No']
counts = [24,7]

plt.figure(figsize=(8, 6));
plt.bar(categories, counts, color='darkgreen')

plt.title('Skipped breakfast before significant event')
plt.xlabel('Choice')
plt.ylabel('Number of Students')

plt.show()
```

```python

labels = ['Yes', 'No', 'Maybe']
sizes = [16, 3, 12]

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['lightgreen', 'lightcoral', 'lightskyblue'])
plt.title('Noticing difference in energy levels')

plt.show()
```
